```python
# Those lines I used to convert in numeric the columns Age and Semester for make the pivot table calculates.
data['Semester'] = pd.to_numeric(data['Semester'], errors='coerce')
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
```

```python
# With this I import the given dataset
file = "E2-Data.csv"
data2 = pd.read_csv(file, sep=";", encoding = 'ANSI')
data2.head()
```

```python
def similarity(data):
    import sys
    import difflib
    results = pd.DataFrame(0, index=data.index, columns=data.index)
    
    # your code in here
    for i, row in data.iterrows():
        A = str(row["Text"])
        for j,row in data.iterrows():
            B = str(row["Text"])
            # I used the library difflib to calculate the ratio of similarity, where 1 are completly equal
            # and 0 both text are completly differents. A number beetwen 0 and 1 is the grade of similarity.
            C = difflib.SequenceMatcher(None, A, B)
            results.iloc[i,j] = C.ratio()
    
    assert results.shape == (data.shape[0], data.shape[0])
    return results
```

```python
# The similarity of example dataset
similarity(data)
```
