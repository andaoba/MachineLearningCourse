# Exercise 13.2

Split the data into training and testing sets

Train a Random Forest Regressor to predict the price of a car using the nominal features

```python
#Import the libraries:
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#Define the data:
nom_df = df.select_dtypes(include=['int64','float64']).copy()
nom_df.head()
```

```python
len(nom_df.columns)
```

```python
#Define the features and the output:
X = nom_df.drop('price',axis=1)
```

```python
y = nom_df['price']
```

```python
# Split the data:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
n_samples = X_train.shape[0]
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
pd.Series(np.sqrt(-cross_val_score(RandomForestRegressor(n_estimators=100), X, y, cv=10, scoring='neg_mean_squared_error'))).describe()
```

# Exercise 13.3

Create dummy variables for the categorical features

Train a Random Forest Regressor and compare

```python
#Redifining X
X = df.drop('price',axis=1)
```

```python
#Get Dummies
X = pd.get_dummies(X)
X.head()
```

```python
# Split the data:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
n_samples = X_train.shape[0]
```

```python
pd.Series(np.sqrt(-cross_val_score(RandomForestRegressor(n_estimators=100), X, y, cv=10, scoring='neg_mean_squared_error'))).describe()
```

This model has a less min RMSE (755.52) compared with the model without the categorical features (896.21)
